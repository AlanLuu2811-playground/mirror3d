```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
from glob import glob
import numpy as np
from sklearn.cluster import KMeans
import json
from pathlib import Path
```

```python
json_path = Path("./mirror3d_input/mp3d/train_10_coarse_normal_mirror.json")

with json_path.open("r") as f:
    data = json.load(f)

def collect_key_values(obj, target_key):
    values = []
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == target_key:
                values.append(v)
            values.extend(collect_key_values(v, target_key))
    elif isinstance(obj, list):
        for item in obj:
            values.extend(collect_key_values(item, target_key))
    return values

mirror_normals = collect_key_values(data, "mirror_normal_camera")

print(f"Found {len(mirror_normals)} values.")
```

```python
X = np.array(mirror_normals, dtype=float)

#X = X[~np.isnan(X).any(axis=1)]

kmeans = KMeans(n_clusters=10)
labels = kmeans.fit_predict(X)

centers = kmeans.cluster_centers_

print("Cluster centers:\n", centers)
print("Inertia:", kmeans.inertia_)
print("Counts per cluster:", np.bincount(labels, minlength=10))
```
